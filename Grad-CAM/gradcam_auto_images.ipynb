```python
import pandas as pd

TARGET_CLASS_NAME = 'CLASS'

df = pd.read_csv('../configs/NIH/test.csv')

label_cols = df.columns.drop("Image Index")

df_class = df[
    (df[TARGET_CLASS_NAME] == 1) &
    (df[label_cols].sum(axis=1) == 1)
]

images_idxs = df_class["Image Index"].tolist()

with open(f"{TARGET_CLASS_NAME.upper()}_IMAGES.txt", "w") as f:
    for img in images_idxs:
        f.write(img + "\n")

dataset_path = '/path/to/CheX-ray14/images/'
gradcam_images_path = f'/path/to/Grad-CAM-images/{TARGET_CLASS_NAME}'

print(f'rsync -av --files-from={TARGET_CLASS_NAME.upper()}_IMAGES.txt {dataset_path} {gradcam_images_path}')
```
